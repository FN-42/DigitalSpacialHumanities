```python
import pandas as pd
```

```python
earliest_date = "2025-06-03" #_13-15-00
int_earliest_date = int(earliest_date.replace("-", ""))*10000
```

## Filter & merge weather data (after earliest data)

```python
# Stationen
# Stations_id von_datum bis_datum Stationshoehe geoBreite geoLaenge Stationsname Bundesland Abgabe
# 02928 19970206 20250625            138     51.3151   12.4462 Leipzig-Holzhausen                       Sachsen                                  Frei                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                     
# 02932 19911101 20250625            131     51.4347   12.2396 Leipzig/Halle                            Sachsen                                  Frei
```

```python
# read in
temp_2928_df = pd.read_csv("Wetterdaten/2928_Leipzig-Holzhausen/temperatur/produkt_zehn_min_tu_20231223_20250624_02928.txt", sep=";", header=0).drop(columns=["  QN", "eor"])
wind_2928_df = pd.read_csv("Wetterdaten/2928_Leipzig-Holzhausen/wind/produkt_zehn_min_ff_20231223_20250624_02928.txt", sep=";", header=0).drop(columns=["  QN", "eor"])
temp_2932_df = pd.read_csv("Wetterdaten/2932_Leipzig-Halle/temperatur/produkt_zehn_min_tu_20231223_20250624_02932.txt", sep=";", header=0).drop(columns=["  QN", "eor"])
wind_2932_df = pd.read_csv("Wetterdaten/2932_Leipzig-Halle/wind/produkt_zehn_min_ff_20231223_20250624_02932.txt", sep=";", header=0).drop(columns=["  QN", "eor"])
```

```python
# filter by earliest date
temp_2928_df = temp_2928_df[temp_2928_df["MESS_DATUM"]>=202505291700].reset_index(drop=True) # (temp_2928_df["MESS_DATUM"]>=int_earliest_date)
wind_2928_df = wind_2928_df[wind_2928_df["MESS_DATUM"]>=202505291700].reset_index(drop=True) # (wind_2928_df["MESS_DATUM"]>=int_earliest_date)
temp_2932_df = temp_2932_df[temp_2932_df["MESS_DATUM"]>=202505291700].reset_index(drop=True) # (temp_2932_df["MESS_DATUM"]>=int_earliest_date)
wind_2932_df = wind_2932_df[wind_2932_df["MESS_DATUM"]>=202505291700].reset_index(drop=True) # (wind_2932_df["MESS_DATUM"]>=int_earliest_date)
```

```python
# Temperaturdaten
# PP_10 [hpa] = Luftdruck in Stationshoehe der voran. 10 min
# TT_10 [°C] = momentane Lufttemperatur in 2m Hoehe
# TM5_10 [°C] = Momentane Temperatur in 5 cm Hoehe 10min
# RF_10 [%] = relative Feucht in 2m Hoehe
# TD_10 [°C] = Taupunkttemperatur in 2m Hoehe

# Winddaten
# FF_10 = mittlere Windgeschwindigkeit der vorangegangenen 10 Minuten [Fehlwert=-999] [m/s]
# DD_10 = mittlere Windrichtung der vorangegangenen 10 Minuten [Fehlwert=-999] [°]
```

```python
all_weather_data = pd.concat(
    [
        pd.merge(temp_2928_df, wind_2928_df, on=["STATIONS_ID", "MESS_DATUM"], how="inner"),
        pd.merge(temp_2932_df, wind_2932_df, on=["STATIONS_ID", "MESS_DATUM"], how="inner"),
    ]
)
all_weather_data.to_csv("Wetterdaten/20250624-0529_all_weather_data.csv", index=False)
all_weather_data
```
